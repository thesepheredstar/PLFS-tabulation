Age Brackets = (0-4, 5-14, 15-29), (15-59), (15-above), (0-4, 5-9, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40-44, 45-49, 50-54, 55-59, 60-64, 65 & above)
Number of years in formal education = (1-4, 5-8, 9-10, 11-12. 13-17, greater than 18)
usual status = employed, unemployed, otside labour force

```python
# createing a column for age bracket 0-29 
age_bins_0_29 = [0, 5, 15, 30, float('inf')]  # Define age ranges
age_labels_0_29 = ['0-4 years', '5-14 years', '15-29 years', '30 years and above']  # Corresponding labels

df['Age_bracket (0-29)'] = pd.cut(df['Age'], bins=age_bins_0_29, labels=age_labels_0_29, right = False)

# creating a column for age bracket 15-29  
age_bins_15_59 = [0, 15, 60, float('inf')]  # Define age ranges
age_labels_15_59 = ['0-14 years', '15-59 years', '60 years and above']  # Corresponding labels

df['Age_bracket(15-59)'] = pd.cut(df['Age'], bins = age_bins_15_59, labels = age_labels_15_59, right = False)

# creating a column for age bracket 15 and above
age_bins_above_15 = [0, 15, float('inf')]
age_label_above_15 = ['0-14 years', '15 years and above']

df['Age_bracket(15 and above)'] = pd.cut(df['Age'], bins = age_bins_above_15, labels = age_label_above_15, right = False)

# creating a column for age bracket 0-65
age_bins_0_65 = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, float('inf')]
age_label_0_65 = ('0-4 years', '5-9 years', '10-14 years', '15-19 years', '20-24 years', '25-29 years', '30-34 years', '35-39 years', '40-44 years', '45-49 years', '50-54 years', '55-59 years', '60-64 years', '65 years & above')

df['age_bracket(0-65)'] = pd.cut(df['Age'], bins = age_bins_0_65, labels = age_label_0_65, right = False)
```

# Tabulation

```python
df_rural = df[df['Sector'] == 'Rural']
df_urban = df[df['Sector'] == 'Urban']
```

### Table 1 (Rural) (df_rt1): Number of villages/blocks, households surveyed and persons enumerated by age and gender for each State

```python
from openpyxl import Workbook

#Table1: rural: column 1

df_rt1_c1 = df_rural.groupby('State')['FSU'].nunique().rename('Number of surveyed villages').reset_index()
df_rt1_c1.loc[len(df_rt1_c1)] = ('All India', df_rt1_c1['Number of surveyed villages'].sum())
df_rt1_c1 = df_rt1_c1.set_index('State')
df_rt1_c1.columns = pd.MultiIndex.from_product([[' '], df_rt1_c1.columns])

#Table1: rural: column 2

df_rt1_c2_sf = df_rural.groupby(['State', 'NSS-Region', 'Stratum','Sub-Stratum','Sub-Sample','FSU','Sample Sg/Sb No.','Second Stage Stratum No.'])['Sample Household Number'].nunique().reset_index()
df_rt1_c2 = df_rt1_c2_sf.groupby('State')['Sample Household Number'].sum().rename('Households').reset_index()
df_rt1_c2.loc[len(df_rt1_c2)] = ('All India', df_rt1_c2['Households'].sum())
df_rt1_c2 = df_rt1_c2.set_index('State')
df_rt1_c2.columns = pd.MultiIndex.from_product([[' '], df_rt1_c2.columns])

#Table1: rural: column 3

rt1_c3_1 = df_rural.groupby(['State',  'Sex','Age_bracket (0-29)']).size().unstack()
rt1_c3_summary_total = rt1_c3_1.groupby('State').sum()
rt1_c3_summary_total['Sex'] = 'Persons'  # Mark the row as Total
rt1_c3_summary_total.set_index('Sex', append=True, inplace=True)
rt1_c3 = pd.concat([rt1_c3_1, rt1_c3_summary_total]).sort_index(level=0)
rt1_c3 = rt1_c3.drop(index='Transgender', level=1)
df_rt1_c3 = rt1_c3.unstack(level = 1)
df_rt1_c3.drop(columns = '30 years and above', level = 'Age_bracket (0-29)', inplace = True)
df_rt1_c3_total = df_rt1_c3.sum(axis=0, numeric_only=True)  # Calculate column-wise total
df_rt1_c3_total.name = 'All India'  # Name the new row
df_rt1_c3 = pd.concat([df_rt1_c3, df_rt1_c3_total.to_frame().T])

#Table1: rural: column 4

rt1_c4_1 = df_rural.groupby(['State',  'Sex','Age_bracket(15-59)']).size().unstack()
rt1_c4_summary_total = rt1_c4_1.groupby('State').sum()
rt1_c4_summary_total['Sex'] = 'Persons'  # Mark the row as Total
rt1_c4_summary_total.set_index('Sex', append=True, inplace=True)
rt1_c4 = pd.concat([rt1_c4_1, rt1_c4_summary_total]).sort_index(level=0)
rt1_c4 = rt1_c4.drop(index='Transgender', level=1)
df_rt1_c4 = rt1_c4.unstack(level = 1)
df_rt1_c4.drop(columns = ['0-14 years','60 years and above'], level ='Age_bracket(15-59)', inplace = True)
df_rt1_c4_total = df_rt1_c4.sum(axis=0, numeric_only=True)  
df_rt1_c4_total.name = 'All India'
df_rt1_c4 = pd.concat([df_rt1_c4, df_rt1_c4_total.to_frame().T])

#Table1: rural: column 5

rt1_c5_1 = df_rural.groupby(['State',  'Sex','Age_bracket(15 and above)']).size().unstack()
rt1_c5_summary_total = rt1_c5_1.groupby('State').sum()
rt1_c5_summary_total['Sex'] = 'Persons' 
rt1_c5_summary_total.set_index('Sex', append=True, inplace=True)
rt1_c5 = pd.concat([rt1_c5_1, rt1_c5_summary_total]).sort_index(level=0)
rt1_c5 = rt1_c5.drop(index='Transgender', level=1)
df_rt1_c5 = rt1_c5.unstack(level = 1)
df_rt1_c5.drop(columns = '0-14 years', level = 'Age_bracket(15 and above)', inplace = True)
df_rt1_c5_total = df_rt1_c5.sum(axis=0, numeric_only=True)
df_rt1_c5_total.name = 'All India'
df_rt1_c5 = pd.concat([df_rt1_c5, df_rt1_c5_total.to_frame().T])

#Table1: rural: column 6

df_rt1_c6 = df_rural.groupby(['State',  'Sex']).size().unstack(fill_value = 0)
df_rt1_c6['Persons'] = df_rt1_c6[['Female', 'Male', 'Transgender']].sum(axis = 1)
df_rt1_c6.columns = pd.MultiIndex.from_product([['All_ages'], df_rt1_c6.columns])
df_rt1_c6_total = df_rt1_c6.sum(axis = 0, numeric_only = True)
df_rt1_c6_total.name = 'All India'
df_rt1_c6 = pd.concat([df_rt1_c6, df_rt1_c6_total.to_frame().T])

#concating all the columns to complete the table 1: rural

df_rt1 = pd.concat([df_rt1_c1, df_rt1_c2, df_rt1_c3, df_rt1_c4, df_rt1_c5, df_rt1_c6], axis=1)

#inserting row for column no.
df_rt1_Column_number = [f"({i+2})" for i in range(len(df_rt1.columns))]
df_rt1_insert_position = df_rt1.index.get_loc('Andaman & N. Island')
df_rt1 = pd.concat([df_rt1.iloc[:df_rt1_insert_position], pd.DataFrame([df_rt1_Column_number], index = ['(1)'], columns = df_rt1.columns), df_rt1.iloc[df_rt1_insert_position:]])

#Moveing UTs at the after states 
df_rt1_rows_to_move = df_rt1.loc[['Andaman & N. Island','D & N. Haveli & Daman & Diu','Jammu & Kashmir','Ladakh','Lakshadweep','Puduchery']]
df_rt1.drop(['Andaman & N. Island','D & N. Haveli & Daman & Diu','Jammu & Kashmir','Ladakh','Lakshadweep','Puduchery'], inplace = True)
df_rt1 = pd.concat([df_rt1.loc[:'West Bengal'],df_rt1_rows_to_move, df_rt1.loc['All India':]])

# renameing the necessary columns
df_rt1.rename(columns = {'D & N. Haveli & Daman & Diu': 'Dadra & Nagar Haveli & Daman & Diu'}, inplace = True)

# assigning a name to the index column
df_rt1.index.name = 'State/UT'
```
